## Vẽ biểu đồ cột, histogram để xem xét xu hướng, quy luật, ...

 Cho đoạn code minh họa Biểu đồ thể hiện số bài đăng theo tháng của năm 2020. Hãy thực hiện các yêu cầu cho các câu hỏi dưới đây:

```python
df1 = newdf.groupby('month').size().reset_index(name='number of records')
n = df1['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df1['month'], df1['number of records'], color=c, width=.5)
for i, val in enumerate(df1['number of records'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df1['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện số bài đăng theo tháng của năm 2020", fontsize=22)
plt.ylabel('number of records')
plt.show()
```

* Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)

```python
#Code ở đây
df2 = newdf.groupby('month').size().reset_index(name='price')
n = df2['month'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)
plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df2['month'], df2['price'], color=c, width=.5)
for i, val in enumerate(df2['price'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})
plt.gca().set_xticklabels(df2['month'], rotation=60, horizontalalignment= 'right')
plt.title("Biểu đồ thể hiện giá nhà đất trung bình theo tháng (VND/m2)", fontsize=22)
plt.ylabel('price')
```

* Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)

```python
data_4 = newdf[newdf['month'] == '4']
data_12 = newdf[newdf['month'] == '12']
fig, ax = plt.subplots(figsize=(5, 5))
sns.displot(data=data_4, x="price", kde=True, ax=ax, color = 'Red')
sns.displot(data=data_12, x="price", kde=True, ax=ax, color= 'Blue')
plt.title("Biểu đồ mật độ giá nhà của tháng 4 và tháng 12 (VND/m2)", fontsize=22)
plt.ylabel('price (VND/m2)')
plt.show()
```

* Số lượng bất động sản phân theo số lượng bedrooms

```python
#Code ở đây
df3 = newdf.groupby('bedrooms').size().reset_index(name='number of records')
n = df3['bedrooms'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df3['bedrooms'], df3['number of records'], color=c, width=.5)
for i, val in enumerate(df3['number of records'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df3['bedrooms'], rotation=60, horizontalalignment= 'right')
plt.title("Số lượng bất động sản phân theo số lượng bedrooms", fontsize=22)
plt.ylabel('number of records')
plt.show()
```

* Giá bất động sản trung bình theo số lượng bedrooms

```python
# Code ở đây
df4 = newdf.groupby('bedrooms')['price'].mean().reset_index(name='average price')
n = df4['bedrooms'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df4['bedrooms'], df4['average price'], color=c, width=.5)
for i, val in enumerate(df4['average price'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df4['bedrooms'], rotation=60, horizontalalignment= 'right')
plt.title("Giá bất động sản trung bình theo số lượng bedrooms", fontsize=22)
plt.ylabel('average price')
plt.show()
```

* Giá nhà đất trung bình theo quận (VND/m2)

```python
# Code ở đây
df5 = newdf.groupby('district')['price'].mean().reset_index(name='average price')
n = df5['district'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df5['district'], df5['average price'], color=c, width=.5)
for i, val in enumerate(df5['average price'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df5['district'], rotation=60, horizontalalignment= 'right')
plt.title("Giá nhà đất trung bình theo quận (VND/m2)", fontsize=22)
plt.ylabel('average price')
plt.show()
```

* Giá nhà đất trung bình theo các phường quận Hoàn Kiếm (VND/m2)

```python
# Filter data for Hoan Kiem district
hoan_kiem_df = newdf[newdf['district'] == 'hoan kiem']

# Group by ward and calculate mean price
df6 = hoan_kiem_df.groupby('ward')['price'].mean().reset_index(name='average price')

# Plotting
n = df6['ward'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df6['ward'], df6['average price'], color=c, width=.5)
for i, val in enumerate(df6['average price'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(df6['ward'], rotation=60, horizontalalignment= 'right')
plt.title("Giá nhà đất trung bình theo các phường quận Hoàn Kiếm (VND/m2)", fontsize=22)
plt.ylabel('average price')
plt.show()
```

* Giá nhà đất trung bình theo nhà đầu tư (VND/m2)

```python
# Randomly sample 10 investors
sampled_investors = newdf['investor'].sample(n=20, random_state=1)
sampled_df = newdf[newdf['investor'].isin(sampled_investors)]

# Group by investor and calculate mean price
df8 = sampled_df.groupby('investor')['price'].mean().reset_index(name='average price')

# Plotting
n = df8['investor'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df8['investor'], df8['average price'], color=c, width=.5)

plt.gca().set_xticklabels(df8['investor'], rotation=60, horizontalalignment= 'right')
plt.title("Giá nhà đất trung bình theo nhà đầu tư (VND/m2)", fontsize=22)
plt.ylabel('average price')
plt.show()
```

* Giá nhà đất trung bình theo dự án (VND/m2)

```python
# Randomly sample 10 projects
sampled_projects = newdf['project'].sample(n=25, random_state=1)
sampled_df = newdf[newdf['project'].isin(sampled_projects)]

# Group by project and calculate mean price
df10 = sampled_df.groupby('project')['price'].mean().reset_index(name='average price')

# Plotting
n = df10['project'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(df10['project'], df10['average price'], color=c, width=.5)

plt.gca().set_xticklabels(df10['project'], rotation=60, horizontalalignment= 'right')
plt.title("Giá nhà đất trung bình theo dự án (VND/m2)", fontsize=22)
plt.ylabel('average price')
plt.show()
```

* Giá bất động sản phân bố theo top 10 nhà đầu tư triệu /m2

```python
# Code ở đây
# Calculate average price per square meter for each investor
newdf['price_per_sqm'] = newdf['price'] / newdf['square']

# Group by investor, calculate mean price per sqm, and sort in descending order
top_10_investors = newdf.groupby('investor')['price_per_sqm'].mean().sort_values(ascending=False).head(10).reset_index()

# Plotting
n = top_10_investors['investor'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(top_10_investors['investor'], top_10_investors['price_per_sqm'], color=c, width=.5)
for i, val in enumerate(top_10_investors['price_per_sqm'].values):
    plt.text(i, val, round(float(val),4), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})

plt.gca().set_xticklabels(top_10_investors['investor'], rotation=60, horizontalalignment= 'right')
plt.title("Giá bất động sản phân bố theo top 10 nhà đầu tư triệu /m2", fontsize=22)
plt.ylabel('average price per sqm')
plt.show()
```

* Giá bất động sản phân bố theo dự án

```python
# Group by project and calculate mean price
# Randomly sample 10 projects
sampled_projects = newdf['project'].sample(n=25, random_state=1)
sampled_df = newdf[newdf['project'].isin(sampled_projects)]

project_prices = sampled_df.groupby('project')['price'].apply(list).reset_index(name='prices')

# Plotting
plt.figure(figsize=(16,10), dpi= 80)
for i, row in project_prices.iterrows():
    plt.scatter([row['project']] * len(row['prices']), row['prices'], label=row['project'])

plt.gca().set_xticklabels(project_prices['project'], rotation=60, horizontalalignment= 'right')
plt.title("Giá bất động sản phân bố theo dự án", fontsize=22)
plt.xlabel('Dự án')
plt.ylabel('Giá')
plt.legend()
plt.show()
```

## Vẽ biểu đồ tương quan giữa các trường thuộc tính

* https://vi.wikipedia.org/wiki/H%E1%BB%87_s%E1%BB%91_t%C6%B0%C6%A1ng_quan
* Tương quan giữa cac cặp biến số thực bằng biểu đồ heatmap

```python
# Gợi ý: sns.heatmap(cmap='RdYlGn', ...)
# Select numerical columns
numerical_cols = ['price', 'bedrooms', 'bathrooms', 'square']
numerical_df = newdf[numerical_cols]

# Calculate correlation matrix
corr_matrix = numerical_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', linewidths=0.5)
plt.title("Tương quan giữa các biến số thực")
plt.show()
```
